# original(4class) 分類精度 (交差検証)

```python
import sys
sys.path.append("../codes")

import numpy as np
import pandas as pd
```

```python
from myData_crossValid import Solver
from set_data_folder import make_train_data
```

## エッジ数同程度サブセット

```python
node = "100"
m_list = [2, 4, 6, 8, 10]
p_list = [0.04, 0.08, 0.11, 0.15, 0.19]
# p_list = [0.004, 0.008, 0.012, 0.016, 0.02]
step_list = [200, 384, 564, 736, 900]
# step_list = [1996, 3984, 5964, 7936, 9900]

df = pd.DataFrame(columns=["parameter", "ave_acc", "std_acc"])

cnt = 0
for m, p, step in zip(m_list, p_list, step_list):
    # make train data folder
    p_s = [
        {"kind": "barabasi", "node": [node], "p": [str(m)]},
        {"kind": "noGrowth", "node": [node], "p": [str(step)]},
        {"kind": "noAttach", "node": [node], "p": [str(m)]},
        {"kind": "random", "node": [node], "p": [str(p)]}
    ]
    print(m, p, step)
    
    make_train_data(p_s, "../train_data/net").copy_data()
    
    solver = Solver(n_splits=10, num_batchs=10, num_epochs=10)
    solver.make_datasets()
    valid_accs = solver.Cross_Valid_Train()
    s = pd.Series(["_".join([str(m),str(p),str(step)]), np.average(valid_accs), np.std(valid_accs)],
                          index=df.columns, name=cnt)

    df = df.append(s)
    cnt += 1
    
```

```python
df.to_csv("./paper_result/4class_GCN_degree.csv")
```
